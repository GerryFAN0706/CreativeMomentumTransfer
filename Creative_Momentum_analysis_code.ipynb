```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols

def analyze_rq1(round2_df):
    """

    """
    print("=== RQ1 Analysis: Effect of AI Suggestion Timing ===\n")

    # 1. Analyze Final Creativity Scores
    print("1. Final Creativity Scores Analysis")
    print("-" * 50)

    # Descriptive statistics
    creativity_stats = round2_df.groupby('condition')[['creativity_score', 'novelty_rating',
                                                     'usefulness_rating', 'originality_rating']].agg(['mean', 'std'])
    print("\nDescriptive Statistics:")
    print(creativity_stats)

    # One-way ANOVA
    model = ols('creativity_score ~ C(condition)', data=round2_df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Calculate effect size (partial eta-squared)
    ss_effect = anova_table['sum_sq'][0]
    ss_total = anova_table['sum_sq'].sum()
    eta_squared = ss_effect / ss_total

    print("\nOne-way ANOVA Results:")
    print(f"F({anova_table.index[0]} df) = {anova_table['F'][0]:.3f}")
    print(f"p-value = {anova_table['PR(>F)'][0]:.6f}")
    print(f"Partial eta-squared = {eta_squared:.3f}")

    # Post-hoc Tukey tests
    tukey = pairwise_tukeyhsd(round2_df['creativity_score'], round2_df['condition'])
    print("\nTukey's HSD Results:")
    print(tukey)

    # 2. Analyze Semantic Distance
    print("\n2. Semantic Distance Analysis")
    print("-" * 50)

    # Descriptive statistics for semantic distance
    distance_stats = round2_df.groupby('condition')['semantic_distance'].agg(['mean', 'std'])
    print("\nSemantic Distance by Condition:")
    print(distance_stats)

    # ANOVA for semantic distance
    distance_model = ols('semantic_distance ~ C(condition)', data=round2_df).fit()
    distance_anova = sm.stats.anova_lm(distance_model, typ=2)

    # Effect size for semantic distance
    ss_effect_dist = distance_anova['sum_sq'][0]
    ss_total_dist = distance_anova['sum_sq'].sum()
    eta_squared_dist = ss_effect_dist / ss_total_dist

    print("\nSemantic Distance ANOVA Results:")
    print(f"F({distance_anova.index[0]} df) = {distance_anova['F'][0]:.3f}")
    print(f"p-value = {distance_anova['PR(>F)'][0]:.6f}")
    print(f"Partial eta-squared = {eta_squared_dist:.3f}")

    # Post-hoc tests for semantic distance
    tukey_distance = pairwise_tukeyhsd(round2_df['semantic_distance'], round2_df['condition'])
    print("\nTukey's HSD Results (Semantic Distance):")
    print(tukey_distance)

    # 3. Visualizations
    plt.figure(figsize=(15, 5))

    # Creativity scores boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(x='condition', y='creativity_score', data=round2_df)
    plt.title('Final Creativity Scores by Condition')
    plt.xticks(rotation=45)

    # Semantic distance boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x='condition', y='semantic_distance', data=round2_df)
    plt.title('Semantic Distance by Condition')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('rq1_analysis.png')
    print("\nVisualization saved as 'rq1_analysis.png'")

    # 4. Component Analysis (Novelty, Usefulness, Originality)
    print("\n3. Creativity Components Analysis")
    print("-" * 50)

    components = ['novelty_rating', 'usefulness_rating', 'originality_rating']
    component_results = {}

    for component in components:
        model = ols(f'{component} ~ C(condition)', data=round2_df).fit()
        anova = sm.stats.anova_lm(model, typ=2)
        ss_effect = anova['sum_sq'][0]
        ss_total = anova['sum_sq'].sum()
        eta_squared = ss_effect / ss_total

        component_results[component] = {
            'F_value': anova['F'][0],
            'p_value': anova['PR(>F)'][0],
            'eta_squared': eta_squared
        }

    print("\nComponent-wise ANOVA Results:")
    for component, results in component_results.items():
        print(f"\n{component.replace('_rating', '').title()}:")
        print(f"F = {results['F_value']:.3f}")
        print(f"p = {results['p_value']:.6f}")
        print(f"η²p = {results['eta_squared']:.3f}")

    return {
        'anova_results': anova_table,
        'tukey_results': tukey,
        'semantic_distance_anova': distance_anova,
        'semantic_distance_tukey': tukey_distance,
        'component_results': component_results,
        'effect_size': eta_squared,
        'semantic_effect_size': eta_squared_dist
    }

# Load and analyze data
try:
    round2_df = pd.read_csv('round2_data.csv')
    results_rq1 = analyze_rq1(round2_df)
except FileNotFoundError:
    print("Please ensure round2_data.csv exists in the current directory!")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

def analyze_rq2(round2_df, post_task_df):
    """
    RQ2: Does Labeling (AI-Labeled vs. Unlabeled) Affect Adoption of AI Ideas and Perceived Influence?

    Analyzes:
    1. Self-reported influence differences between labeled/unlabeled
    2. Semantic distance patterns by labeling
    3. Correlation between perceived and actual influence
    """
    print("=== RQ2 Analysis: Effect of AI Labeling ===\n")

    # Merge relevant data
    analysis_df = round2_df.merge(
        post_task_df[['participant_id', 'perceived_external_influence', 'recognized_ai']],
        on='participant_id'
    )

    # Create labeling groups (excluding control)
    analysis_df['labeling'] = analysis_df['condition'].apply(
        lambda x: 'Labeled' if 'Labeled' in x else 'Unlabeled' if 'Unlabeled' in x else 'Control'
    )

    # 1. Analyze Self-Reported Influence
    print("1. Self-Reported Influence Analysis")
    print("-" * 50)

    # Descriptive statistics
    influence_stats = analysis_df[analysis_df['labeling'] != 'Control'].groupby('labeling')[
        'perceived_external_influence'].agg(['mean', 'std', 'count'])
    print("\nDescriptive Statistics (Self-Reported Influence):")
    print(influence_stats)

    # T-test for labeled vs unlabeled
    labeled_influence = analysis_df[analysis_df['labeling'] == 'Labeled']['perceived_external_influence']
    unlabeled_influence = analysis_df[analysis_df['labeling'] == 'Unlabeled']['perceived_external_influence']

    t_stat, p_val = stats.ttest_ind(labeled_influence, unlabeled_influence)
    cohen_d = (labeled_influence.mean() - unlabeled_influence.mean()) / np.sqrt(
        ((labeled_influence.count() - 1) * labeled_influence.var() +
         (unlabeled_influence.count() - 1) * unlabeled_influence.var()) /
        (labeled_influence.count() + unlabeled_influence.count() - 2)
    )

    print("\nt-test Results (Labeled vs. Unlabeled):")
    print(f"t = {t_stat:.3f}")
    print(f"p = {p_val:.6f}")
    print(f"Cohen's d = {cohen_d:.3f}")

    # 2. Analyze Semantic Distance
    print("\n2. Semantic Distance Analysis by Labeling")
    print("-" * 50)

    # Descriptive statistics
    distance_stats = analysis_df[analysis_df['labeling'] != 'Control'].groupby('labeling')[
        'semantic_distance'].agg(['mean', 'std', 'count'])
    print("\nDescriptive Statistics (Semantic Distance):")
    print(distance_stats)

    # T-test for semantic distance
    labeled_distance = analysis_df[analysis_df['labeling'] == 'Labeled']['semantic_distance']
    unlabeled_distance = analysis_df[analysis_df['labeling'] == 'Unlabeled']['semantic_distance']

    t_stat_dist, p_val_dist = stats.ttest_ind(labeled_distance, unlabeled_distance)
    cohen_d_dist = (labeled_distance.mean() - unlabeled_distance.mean()) / np.sqrt(
        ((labeled_distance.count() - 1) * labeled_distance.var() +
         (unlabeled_distance.count() - 1) * unlabeled_distance.var()) /
        (labeled_distance.count() + unlabeled_distance.count() - 2)
    )

    print("\nt-test Results (Semantic Distance):")
    print(f"t = {t_stat_dist:.3f}")
    print(f"p = {p_val_dist:.6f}")
    print(f"Cohen's d = {cohen_d_dist:.3f}")

    # 3. Correlation Analysis
    print("\n3. Correlation Analysis")
    print("-" * 50)

    # Calculate correlation between perceived influence and semantic distance
    non_control_df = analysis_df[analysis_df['labeling'] != 'Control']
    corr_coef, corr_p = pearsonr(
        non_control_df['perceived_external_influence'],
        non_control_df['semantic_distance']
    )

    print("\nCorrelation between Perceived Influence and Semantic Distance:")
    print(f"r = {corr_coef:.3f}")
    print(f"p = {corr_p:.6f}")

    # 4. AI Recognition Analysis (for unlabeled conditions)
    print("\n4. AI Recognition Analysis")
    print("-" * 50)

    unlabeled_recognition = analysis_df[
        analysis_df['labeling'] == 'Unlabeled'
    ]['recognized_ai'].value_counts(normalize=True)

    print("\nProportion of Participants who Recognized AI in Unlabeled Conditions:")
    print(unlabeled_recognition)

    # 5. Visualizations
    plt.figure(figsize=(15, 5))

    # Self-reported influence by labeling
    plt.subplot(1, 3, 1)
    sns.boxplot(x='labeling', y='perceived_external_influence',
                data=analysis_df[analysis_df['labeling'] != 'Control'])
    plt.title('Self-Reported Influence by Labeling')
    plt.xticks(rotation=45)

    # Semantic distance by labeling
    plt.subplot(1, 3, 2)
    sns.boxplot(x='labeling', y='semantic_distance',
                data=analysis_df[analysis_df['labeling'] != 'Control'])
    plt.title('Semantic Distance by Labeling')
    plt.xticks(rotation=45)

    # Correlation scatter plot
    plt.subplot(1, 3, 3)
    sns.scatterplot(data=non_control_df, x='semantic_distance',
                    y='perceived_external_influence', hue='labeling', alpha=0.6)
    plt.title('Perceived Influence vs. Semantic Distance')

    plt.tight_layout()
    plt.savefig('rq2_analysis.png')
    print("\nVisualization saved as 'rq2_analysis.png'")

    return {
        'influence_stats': influence_stats,
        'distance_stats': distance_stats,
        'influence_ttest': {'t': t_stat, 'p': p_val, 'cohen_d': cohen_d},
        'distance_ttest': {'t': t_stat_dist, 'p': p_val_dist, 'cohen_d': cohen_d_dist},
        'correlation': {'r': corr_coef, 'p': corr_p},
        'recognition_rates': unlabeled_recognition
    }

# Load and analyze data
try:
    round2_df = pd.read_csv('round2_data.csv')
    post_task_df = pd.read_csv('post_task_data.csv')
    results_rq2 = analyze_rq2(round2_df, post_task_df)
except FileNotFoundError:
    print("Please ensure both round2_data.csv and post_task_data.csv exist!")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_rq3(round2_df, prestudy_df):
    """
    RQ3: Do Individual Differences (Baseline Creativity, Trust in AI)
    Moderate the Effect of AI Suggestions on Final Outcomes?
    """
    print("=== RQ3 Analysis: Moderation Effects ===\n")

    # Merge data
    analysis_df = round2_df.merge(
        prestudy_df[['participant_id', 'aut_fluency', 'trust_mean']],
        on='participant_id'
    )

    # Center continuous variables
    analysis_df['aut_fluency_centered'] = analysis_df['aut_fluency'] - analysis_df['aut_fluency'].mean()
    analysis_df['trust_mean_centered'] = analysis_df['trust_mean'] - analysis_df['trust_mean'].mean()

    # Create binary indicators for high/low groups (for visualization)
    analysis_df['high_creativity'] = analysis_df['aut_fluency'] > analysis_df['aut_fluency'].median()
    analysis_df['high_trust'] = analysis_df['trust_mean'] > analysis_df['trust_mean'].median()

    # 1. Baseline Creativity Moderation
    print("1. Baseline Creativity Moderation Analysis")
    print("-" * 50)

    # Model with creativity interaction
    creativity_model = ols(
        'creativity_score ~ C(condition) + aut_fluency_centered + C(condition):aut_fluency_centered',
        data=analysis_df
    ).fit()

    print("\nModeration Analysis (Baseline Creativity):")
    print(creativity_model.summary().tables[1])

    # Calculate effect sizes
    creativity_anova = sm.stats.anova_lm(creativity_model, typ=2)
    print("\nANOVA Results (Creativity Moderation):")
    print(creativity_anova)

    # 2. Trust in AI Moderation
    print("\n2. Trust in AI Moderation Analysis")
    print("-" * 50)

    # Model with trust interaction
    trust_model = ols(
        'creativity_score ~ C(condition) + trust_mean_centered + C(condition):trust_mean_centered',
        data=analysis_df
    ).fit()

    print("\nModeration Analysis (Trust in AI):")
    print(trust_model.summary().tables[1])

    # Calculate effect sizes
    trust_anova = sm.stats.anova_lm(trust_model, typ=2)
    print("\nANOVA Results (Trust Moderation):")
    print(trust_anova)

    # 3. Simple Slopes Analysis
    print("\n3. Simple Slopes Analysis")
    print("-" * 50)

    # Function to compute simple slopes
    def compute_simple_slopes(model, moderator_name, conditions):
        slopes = {}
        for condition in conditions:
            mask = analysis_df['condition'] == condition
            slope, _, _, _, _ = stats.linregress(
                analysis_df.loc[mask, f'{moderator_name}_centered'],
                analysis_df.loc[mask, 'creativity_score']
            )
            slopes[condition] = slope
        return slopes

    creativity_slopes = compute_simple_slopes(
        creativity_model, 'aut_fluency',
        analysis_df['condition'].unique()
    )

    trust_slopes = compute_simple_slopes(
        trust_model, 'trust_mean',
        analysis_df['condition'].unique()
    )

    print("\nCreativity Simple Slopes by Condition:")
    for condition, slope in creativity_slopes.items():
        print(f"{condition}: {slope:.3f}")

    print("\nTrust Simple Slopes by Condition:")
    for condition, slope in trust_slopes.items():
        print(f"{condition}: {slope:.3f}")

    # 4. Visualizations
    plt.figure(figsize=(15, 10))

    # Creativity moderation plot
    plt.subplot(2, 2, 1)
    sns.scatterplot(
        data=analysis_df,
        x='aut_fluency_centered',
        y='creativity_score',
        hue='condition',
        alpha=0.6
    )
    plt.title('Creativity Score vs. Baseline Creativity')
    plt.xlabel('Baseline Creativity (Centered)')

    # Trust moderation plot
    plt.subplot(2, 2, 2)
    sns.scatterplot(
        data=analysis_df,
        x='trust_mean_centered',
        y='creativity_score',
        hue='condition',
        alpha=0.6
    )
    plt.title('Creativity Score vs. Trust in AI')
    plt.xlabel('Trust in AI (Centered)')

    # Box plots for high/low groups
    plt.subplot(2, 2, 3)
    sns.boxplot(
        data=analysis_df,
        x='condition',
        y='creativity_score',
        hue='high_creativity'
    )
    plt.title('Creativity Scores by Condition and Baseline Creativity')
    plt.xticks(rotation=45)

    plt.subplot(2, 2, 4)
    sns.boxplot(
        data=analysis_df,
        x='condition',
        y='creativity_score',
        hue='high_trust'
    )
    plt.title('Creativity Scores by Condition and Trust in AI')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('rq3_analysis.png')
    print("\nVisualization saved as 'rq3_analysis.png'")

    # 5. Additional Analyses: Semantic Distance Moderation
    print("\n4. Semantic Distance Moderation")
    print("-" * 50)

    # Only analyze AI conditions for semantic distance
    ai_conditions_df = analysis_df[analysis_df['condition'].str.contains('AI')]

    # Creativity moderation on semantic distance
    sem_dist_creativity_model = ols(
        'semantic_distance ~ C(condition) + aut_fluency_centered + C(condition):aut_fluency_centered',
        data=ai_conditions_df
    ).fit()

    # Trust moderation on semantic distance
    sem_dist_trust_model = ols(
        'semantic_distance ~ C(condition) + trust_mean_centered + C(condition):trust_mean_centered',
        data=ai_conditions_df
    ).fit()

    print("\nCreativity Moderation on Semantic Distance:")
    print(sem_dist_creativity_model.summary().tables[1])

    print("\nTrust Moderation on Semantic Distance:")
    print(sem_dist_trust_model.summary().tables[1])

    return {
        'creativity_moderation': creativity_model,
        'trust_moderation': trust_model,
        'creativity_slopes': creativity_slopes,
        'trust_slopes': trust_slopes,
        'semantic_creativity_mod': sem_dist_creativity_model,
        'semantic_trust_mod': sem_dist_trust_model
    }

# Load and analyze data
try:
    round2_df = pd.read_csv('round2_data.csv')
    prestudy_df = pd.read_csv('prestudy_data.csv')
    results_rq3 = analyze_rq3(round2_df, prestudy_df)
except FileNotFoundError:
    print("Please ensure both round2_data.csv and prestudy_data.csv exist!")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

def analyze_rq4(round1_df, round2_df):
    """
    RQ4: How Does Creative Momentum Transfer Evolve Across Rounds?

    Analyzes:
    1. Within-subject changes in creativity scores
    2. Evolution of semantic distance
    3. Condition-specific trajectories
    """
    print("=== RQ4 Analysis: Creative Momentum Transfer Evolution ===\n")

    # Merge round1 and round2 data
    analysis_df = round1_df.merge(
        round2_df,
        on=['participant_id', 'condition'],
        suffixes=('_r1', '_r2')
    )

    # Calculate change scores
    analysis_df['creativity_change'] = analysis_df['creativity_score_r2'] - analysis_df['creativity_score_r1']
    analysis_df['novelty_change'] = analysis_df['novelty_rating_r2'] - analysis_df['novelty_rating_r1']
    analysis_df['usefulness_change'] = analysis_df['usefulness_rating_r2'] - analysis_df['usefulness_rating_r1']
    analysis_df['originality_change'] = analysis_df['originality_rating_r2'] - analysis_df['originality_rating_r1']
    analysis_df['semantic_distance_change'] = analysis_df['semantic_distance_r2'] - analysis_df['semantic_distance_r1']

    # 1. Overall Change Analysis
    print("1. Overall Change Analysis")
    print("-" * 50)

    # Descriptive statistics for changes by condition
    change_stats = analysis_df.groupby('condition')[
        ['creativity_change', 'semantic_distance_change']
    ].agg(['mean', 'std'])
    print("\nChange Statistics by Condition:")
    print(change_stats)

    # Repeated measures ANOVA
    model = sm.OLS.from_formula('creativity_change ~ C(condition)', data=analysis_df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("\nANOVA Results for Creativity Change:")
    print(anova_table)

    # 2. Late-AI Pivot Analysis
    print("\n2. Late-AI Pivot Analysis")
    print("-" * 50)

    # Analyze specifically Late-AI conditions
    late_ai_df = analysis_df[analysis_df['condition'].str.contains('Late-AI')]

    # T-test for significant change in Late-AI conditions
    t_stat, p_val = stats.ttest_rel(
        late_ai_df['creativity_score_r1'],
        late_ai_df['creativity_score_r2']
    )

    print("\nLate-AI Conditions Change (Paired t-test):")
    print(f"t = {t_stat:.3f}")
    print(f"p = {p_val:.6f}")

    # Effect size (Cohen's d for paired samples)
    d = (late_ai_df['creativity_score_r2'].mean() - late_ai_df['creativity_score_r1'].mean()) / \
        np.sqrt((late_ai_df['creativity_score_r2'].std()**2 + late_ai_df['creativity_score_r1'].std()**2) / 2)
    print(f"Cohen's d = {d:.3f}")

    # 3. Component-wise Analysis
    print("\n3. Component-wise Changes")
    print("-" * 50)

    components = ['novelty_change', 'usefulness_change', 'originality_change']

    for component in components:
        model = sm.OLS.from_formula(f'{component} ~ C(condition)', data=analysis_df).fit()
        anova = sm.stats.anova_lm(model, typ=2)
        print(f"\n{component.replace('_change', '')} Change ANOVA:")
        print(anova)

    # 4. Semantic Distance Evolution
    print("\n4. Semantic Distance Evolution")
    print("-" * 50)

    # Analyze semantic distance changes
    dist_model = sm.OLS.from_formula('semantic_distance_change ~ C(condition)', data=analysis_df).fit()
    dist_anova = sm.stats.anova_lm(dist_model, typ=2)

    print("\nSemantic Distance Change ANOVA:")
    print(dist_anova)

    # 5. Visualizations
    plt.figure(figsize=(15, 10))

    # Creativity score evolution
    plt.subplot(2, 2, 1)
    for condition in analysis_df['condition'].unique():
        condition_data = analysis_df[analysis_df['condition'] == condition]
        plt.plot(
            [1, 2],
            [condition_data['creativity_score_r1'].mean(),
             condition_data['creativity_score_r2'].mean()],
            'o-',
            label=condition
        )
    plt.title('Evolution of Creativity Scores')
    plt.xlabel('Round')
    plt.ylabel('Mean Creativity Score')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)

    # Box plot of changes
    plt.subplot(2, 2, 2)
    sns.boxplot(
        data=analysis_df,
        x='condition',
        y='creativity_change'
    )
    plt.title('Creativity Score Changes by Condition')
    plt.xticks(rotation=45)

    # Semantic distance evolution
    plt.subplot(2, 2, 3)
    for condition in analysis_df['condition'].unique():
        condition_data = analysis_df[analysis_df['condition'] == condition]
        plt.plot(
            [1, 2],
            [condition_data['semantic_distance_r1'].mean(),
             condition_data['semantic_distance_r2'].mean()],
            'o-',
            label=condition
        )
    plt.title('Evolution of Semantic Distance')
    plt.xlabel('Round')
    plt.ylabel('Mean Semantic Distance')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)

    # Component changes
    plt.subplot(2, 2, 4)
    changes_df = pd.melt(
        analysis_df,
        id_vars=['condition'],
        value_vars=components,
        var_name='component',
        value_name='change'
    )
    sns.boxplot(
        data=changes_df,
        x='condition',
        y='change',
        hue='component'
    )
    plt.title('Component Changes by Condition')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('rq4_analysis.png')
    print("\nVisualization saved as 'rq4_analysis.png'")

    return {
        'change_stats': change_stats,
        'anova_results': anova_table,
        'late_ai_ttest': {'t': t_stat, 'p': p_val, 'cohens_d': d},
        'semantic_distance_anova': dist_anova
    }

# Load and analyze data
try:
    round1_df = pd.read_csv('round1_data.csv')
    round2_df = pd.read_csv('round2_data.csv')
    results_rq4 = analyze_rq4(round1_df, round2_df)
except FileNotFoundError:
    print("Please ensure both round1_data.csv and round2_data.csv exist!")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

def create_pivot_visualization(round1_df, round2_df):
    """
    Create enhanced visualization of the pivot effect using raw data
    and multiple visualization techniques
    """
    # Merge round1 and round2 data
    analysis_df = round1_df.merge(
        round2_df,
        on=['participant_id', 'condition'],
        suffixes=('_r1', '_r2')
    )

    # Calculate changes
    analysis_df['creativity_change'] = analysis_df['creativity_score_r2'] - analysis_df['creativity_score_r1']
    analysis_df['semantic_change'] = analysis_df['semantic_distance_r2'] - analysis_df['semantic_distance_r1']

    # Create figure with multiple subplots
    fig = plt.figure(figsize=(20, 15))

    # Color scheme
    colors = {
        'Control': '#808080',
        'Early-AI-Labeled': '#1f77b4',
        'Early-AI-Unlabeled': '#17becf',
        'Late-AI-Labeled': '#d62728',
        'Late-AI-Unlabeled': '#ff7f0e'
    }

    # 1. Spaghetti Plot with Individual Trajectories
    ax1 = plt.subplot(221)

    # Plot individual trajectories with transparency
    for condition in analysis_df['condition'].unique():
        condition_data = analysis_df[analysis_df['condition'] == condition]

        # Plot individual lines with low alpha
        for _, row in condition_data.iterrows():
            plt.plot([1, 2], [row['creativity_score_r1'], row['creativity_score_r2']],
                    '-', color=colors[condition], alpha=0.1)

        # Plot mean line with high alpha
        mean_r1 = condition_data['creativity_score_r1'].mean()
        mean_r2 = condition_data['creativity_score_r2'].mean()
        plt.plot([1, 2], [mean_r1, mean_r2], 'o-',
                color=colors[condition], label=condition,
                linewidth=3, markersize=10)

    plt.title('Individual Creativity Score Trajectories\n(with condition means)', fontsize=14)
    plt.xlabel('Round', fontsize=12)
    plt.ylabel('Creativity Score', fontsize=12)
    plt.xticks([1, 2], ['Round 1', 'Round 2'])
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(fontsize=10)

    # 2. Slope Plot focusing on Late-AI Pivot
    ax2 = plt.subplot(222)

    late_ai_data = analysis_df[analysis_df['condition'].str.contains('Late-AI')]

    # Create slope plot
    for _, row in late_ai_data.iterrows():
        plt.plot([1, 2], [row['creativity_score_r1'], row['creativity_score_r2']],
                '-', color=colors[row['condition']], alpha=0.3)

    # Add mean lines
    for condition in ['Late-AI-Labeled', 'Late-AI-Unlabeled']:
        condition_data = late_ai_data[late_ai_data['condition'] == condition]
        mean_r1 = condition_data['creativity_score_r1'].mean()
        mean_r2 = condition_data['creativity_score_r2'].mean()
        plt.plot([1, 2], [mean_r1, mean_r2], 'o-',
                color=colors[condition], label=f'{condition} (Mean)',
                linewidth=4, markersize=12)

        # Add arrow to emphasize pivot
        plt.annotate('', xy=(2, mean_r2), xytext=(2, mean_r1),
                    arrowprops=dict(arrowstyle='->',
                                  color=colors[condition],
                                  lw=2,
                                  mutation_scale=20))

        # Add text showing magnitude of change
        plt.text(2.1, (mean_r1 + mean_r2)/2,
                f'Δ = {mean_r2-mean_r1:.2f}',
                color=colors[condition])

    plt.title('Late-AI Pivot Effect Detail', fontsize=14)
    plt.xlabel('Round', fontsize=12)
    plt.ylabel('Creativity Score', fontsize=12)
    plt.xticks([1, 2], ['Pre-AI', 'Post-AI'])
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(fontsize=10)

    # 3. Change Distribution
    ax3 = plt.subplot(223)

    sns.violinplot(data=analysis_df, x='condition', y='creativity_change',
                  palette=colors, inner='box')
    plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
    plt.title('Distribution of Creativity Score Changes', fontsize=14)
    plt.xlabel('Condition', fontsize=12)
    plt.ylabel('Change in Creativity Score', fontsize=12)
    plt.xticks(rotation=45)

    # 4. Paired Scatter Plot
    ax4 = plt.subplot(224)

    # Plot diagonal reference line
    min_score = min(analysis_df['creativity_score_r1'].min(),
                   analysis_df['creativity_score_r2'].min())
    max_score = max(analysis_df['creativity_score_r1'].max(),
                   analysis_df['creativity_score_r2'].max())
    plt.plot([min_score, max_score], [min_score, max_score],
             '--', color='gray', alpha=0.5, label='No Change')

    # Plot scatter points
    for condition in analysis_df['condition'].unique():
        mask = analysis_df['condition'] == condition
        plt.scatter(analysis_df[mask]['creativity_score_r1'],
                   analysis_df[mask]['creativity_score_r2'],
                   label=condition, color=colors[condition], alpha=0.6)

        # Add arrows for mean change
        mean_r1 = analysis_df[mask]['creativity_score_r1'].mean()
        mean_r2 = analysis_df[mask]['creativity_score_r2'].mean()
        if 'Late-AI' in condition:  # Emphasize Late-AI conditions
            plt.arrow(mean_r1, mean_r1, 0, mean_r2-mean_r1,
                     head_width=0.1, head_length=0.1,
                     color=colors[condition], alpha=0.8)

    plt.title('Round 1 vs Round 2 Creativity Scores', fontsize=14)
    plt.xlabel('Round 1 Score', fontsize=12)
    plt.ylabel('Round 2 Score', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1))

    plt.tight_layout()
    plt.savefig('pivot_effect_detailed.png', bbox_inches='tight', dpi=300)
    print("\nDetailed pivot effect visualization saved as 'pivot_effect_detailed.png'")

    # Create summary statistics
    summary_stats = pd.DataFrame({
        'condition': analysis_df['condition'].unique(),
        'mean_change': [analysis_df[analysis_df['condition'] == c]['creativity_change'].mean()
                       for c in analysis_df['condition'].unique()],
        'std_change': [analysis_df[analysis_df['condition'] == c]['creativity_change'].std()
                      for c in analysis_df['condition'].unique()],
        'effect_size': [analysis_df[analysis_df['condition'] == c]['creativity_change'].mean() /
                       analysis_df[analysis_df['condition'] == c]['creativity_change'].std()
                       for c in analysis_df['condition'].unique()]
    })

    return summary_stats

# Load and create visualization
try:
    round1_df = pd.read_csv('round1_data.csv')
    round2_df = pd.read_csv('round2_data.csv')
    summary_stats = create_pivot_visualization(round1_df, round2_df)
    print("\nSummary Statistics for Changes:")
    print(summary_stats)
except FileNotFoundError:
    print("Please ensure both round1_data.csv and round2_data.csv exist!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def plot_baseline_creativity_moderation(analysis_df):
    # Set style


    # Create figure
    plt.figure(figsize=(10, 6))

    # Color palette for conditions
    colors = {
        'Control': '#808080',
        'Early-AI-Labeled': '#1f77b4',
        'Early-AI-Unlabeled': '#2ca02c',
        'Late-AI-Labeled': '#ff7f0e',
        'Late-AI-Unlabeled': '#d62728'
    }

    # Plot scatter points and regression lines for each condition
    for condition in analysis_df['condition'].unique():
        mask = analysis_df['condition'] == condition
        condition_data = analysis_df[mask]

        # Scatter plot
        plt.scatter(condition_data['aut_fluency_centered'],
                   condition_data['creativity_score'],
                   alpha=0.3,
                   color=colors[condition],
                   label=condition)

        # Regression line
        x = condition_data['aut_fluency_centered']
        y = condition_data['creativity_score']

        # Calculate regression
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

        # Generate points for regression line
        x_reg = np.linspace(x.min(), x.max(), 100)
        y_reg = slope * x_reg + intercept

        # Calculate confidence intervals
        n = len(x)
        mean_x = np.mean(x)
        se_y = std_err * np.sqrt(1/n + (x_reg - mean_x)**2 / np.sum((x - mean_x)**2))

        # Plot regression line and confidence interval
        plt.plot(x_reg, y_reg, '-', color=colors[condition], linewidth=2)
        plt.fill_between(x_reg,
                        y_reg - 1.96 * se_y,
                        y_reg + 1.96 * se_y,
                        color=colors[condition],
                        alpha=0.1)

    # Customize plot
    plt.xlabel('Baseline Creativity (Centered)', fontsize=12)
    plt.ylabel('Final Creativity Score', fontsize=12)
    plt.title('Baseline Creativity Moderation Effect by Condition', fontsize=14)

    # Add legend
    plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Add grid
    plt.grid(True, alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Add text box with key statistics
    stats_text = f'Interaction Effect:\nF = 4.42\np = .002'
    plt.text(0.02, 0.98, stats_text,
             transform=plt.gca().transAxes,
             bbox=dict(facecolor='white', alpha=0.8),
             verticalalignment='top')

    return plt

# Example usage with your data:
analysis_df = pd.DataFrame({
    'condition': round2_df['condition'],
    'aut_fluency_centered': prestudy_df['aut_fluency'] - prestudy_df['aut_fluency'].mean(),
    'creativity_score': round2_df['creativity_score']
})

# Create and save the plot
plot = plot_baseline_creativity_moderation(analysis_df)
plot.savefig('baseline_creativity_moderation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_creativity_evolution(round1_df, round2_df):
    # Set style

    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Color palette
    colors = {
        'Control': '#808080',
        'Early-AI-Labeled': '#1f77b4',
        'Early-AI-Unlabeled': '#2ca02c',
        'Late-AI-Labeled': '#ff7f0e',
        'Late-AI-Unlabeled': '#d62728'
    }

    # Plot 1: Creativity Score Evolution
    for condition in round1_df['condition'].unique():
        mask = round1_df['condition'] == condition
        # Calculate means for each round
        r1_mean = round1_df[mask]['creativity_score'].mean()
        r2_mean = round2_df[mask]['creativity_score'].mean()
        # Calculate standard errors
        r1_se = round1_df[mask]['creativity_score'].std() / np.sqrt(sum(mask))
        r2_se = round2_df[mask]['creativity_score'].std() / np.sqrt(sum(mask))

        # Plot lines
        ax1.plot([1, 2], [r1_mean, r2_mean], '-o',
                color=colors[condition], label=condition, linewidth=2)
        # Add error bars
        ax1.fill_between([1, 2],
                        [r1_mean - r1_se, r2_mean - r2_se],
                        [r1_mean + r1_se, r2_mean + r2_se],
                        color=colors[condition], alpha=0.2)

    ax1.set_xlabel('Round', fontsize=12)
    ax1.set_ylabel('Creativity Score', fontsize=12)
    ax1.set_title('Evolution of Creativity Scores', fontsize=14)
    ax1.set_xticks([1, 2])
    ax1.legend(title='Condition', bbox_to_anchor=(1.05, 1))

    # Plot 2: Semantic Distance Evolution
    for condition in round1_df['condition'].unique():
        if condition != 'Control':  # Skip control for semantic distance
            mask = round1_df['condition'] == condition
            # Calculate means for each round
            r1_mean = round1_df[mask]['semantic_distance'].mean()
            r2_mean = round2_df[mask]['semantic_distance'].mean()
            # Calculate standard errors
            r1_se = round1_df[mask]['semantic_distance'].std() / np.sqrt(sum(mask))
            r2_se = round2_df[mask]['semantic_distance'].std() / np.sqrt(sum(mask))

            # Plot lines
            ax2.plot([1, 2], [r1_mean, r2_mean], '-o',
                    color=colors[condition], label=condition, linewidth=2)
            # Add error bars
            ax2.fill_between([1, 2],
                           [r1_mean - r1_se, r2_mean - r2_se],
                           [r1_mean + r1_se, r2_mean + r2_se],
                           color=colors[condition], alpha=0.2)

    ax2.set_xlabel('Round', fontsize=12)
    ax2.set_ylabel('Semantic Distance', fontsize=12)
    ax2.set_title('Evolution of Semantic Distance', fontsize=14)
    ax2.set_xticks([1, 2])
    ax2.legend(title='Condition', bbox_to_anchor=(1.05, 1))

    # Add overall title
    fig.suptitle('Creative Momentum Transfer Across Rounds', fontsize=16, y=1.05)

    # Adjust layout
    plt.tight_layout()

    return fig

# Example usage:
fig = plot_creativity_evolution(round1_df, round2_df)
plt.savefig('rq4_evolution.png', dpi=300, bbox_inches='tight')
plt.show()
```
